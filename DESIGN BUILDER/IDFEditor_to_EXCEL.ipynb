```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

def idf_constructions_to_excel(input_string: str, output_file: str, group_size: int = 5):
    """
    Convertit une chaîne IDF Constructions en Excel avec un onglet par groupe de `group_size` constructions.
    - Chaque onglet contient les propriétés transposées (Field en lignes, ObjX en colonnes)
    - Les constructions contenant '_Rev' sont ignorées
    - Police Arial, colonnes auto-ajustées
    """
    # Découper chaque construction
    rows = [r.strip() for r in input_string.strip().split(';') if r.strip()]

    constructions = []
    for row in rows:
        parts = [p.strip() for p in row.split(',')]
        if not parts[0].lower().startswith("construction"):
            continue
        name = parts[1]
        if "_Rev" in name:  # ignorer les versions inversées
            continue
        layers = parts[2:]
        cons = {"Name": name}
        for i, layer in enumerate(layers):
            if layer:
                if i == 0:
                    cons["Outside Layer"] = layer
                else:
                    cons[f"Layer {i+1}"] = layer
        constructions.append(cons)

    # Liste de tous les champs possibles
    max_layers = max(len(c) for c in constructions)
    fields = ["Name", "Outside Layer"] + [f"Layer {i}" for i in range(2, max_layers)]

    # Ecriture en onglets
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        total = len(constructions)
        for start in range(0, total, group_size):
            block = constructions[start:start+group_size]
            data = {f: [] for f in fields}
            for cons in block:
                for f in fields:
                    data[f].append(cons.get(f, ""))
            obj_labels = [f"Obj{j+1}" for j in range(start, start+len(block))]
            df = pd.DataFrame(data, index=obj_labels).T
            df.index.name = "Field"
            sheet_name = f"Obj{start+1}-{start+len(block)}"
            df.to_excel(writer, sheet_name=sheet_name)

    # Mise en forme
    wb = load_workbook(output_file)
    for ws in wb.worksheets:
        for row in ws.iter_rows():
            for cell in row:
                cell.font = Font(name="Arial", size=10)
        for col in ws.columns:
            max_len = 0
            col_letter = col[0].column_letter
            for cell in col:
                if cell.value:
                    max_len = max(max_len, len(str(cell.value)))
            ws.column_dimensions[col_letter].width = max_len + 2
    wb.save(output_file)
    print(f"✅ Excel généré avec {len(constructions)} constructions regroupées par {group_size} par onglet : {output_file}")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

def idf_materials_to_excel(input_string: str, output_file: str, group_size: int = 5):
    """
    Parse les objets 'Material' d'un bloc IDF et exporte vers Excel.
    Champs extraits : Name, Roughness, Thickness, Conductivity, Density, Specific Heat,
                      Thermal Absorptance, Solar Absorptance, Visible Absorptance.
    - Gère les lignes avec préfixe 'IDF,'.
    - Regroupe par groupes de `group_size` dans des onglets séparés.
    - Mise en forme : police Arial et colonnes auto-ajustées.
    """
    KEY = "Material"

    # Split et parse
    rows = [r.strip() for r in input_string.strip().split(";") if r.strip()]
    materials = []

    for row in rows:
        parts = [p.strip() for p in row.split(",")]

        # Cas où la ligne commence par "IDF,"
        if parts[0].upper() == "IDF":
            parts = parts[1:]

        if not parts or parts[0] != KEY:
            continue

        if len(parts) < 10:
            continue  # ligne incomplète

        materials.append({
            "Name": parts[1],
            "Roughness": parts[2],
            "Thickness (m)": parts[3],
            "Conductivity (W/mK)": parts[4],
            "Density (kg/m3)": parts[5],
            "Specific Heat (J/kgK)": parts[6],
            "Thermal Absorptance": parts[7],
            "Solar Absorptance": parts[8],
            "Visible Absorptance": parts[9],
        })

    # Export Excel avec regroupement
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for start in range(0, len(materials), group_size):
            block = materials[start:start + group_size]
            df = pd.DataFrame(block, index=[f"Obj{i+1}" for i in range(start, start+len(block))]).T
            df.index.name = "Field"
            sheet_name = f"Obj{start+1}-{start+len(block)}"
            df.to_excel(writer, sheet_name=sheet_name)

    # Mise en forme
    wb = load_workbook(output_file)
    for ws in wb.worksheets:
        for row in ws.iter_rows():
            for cell in row:
                cell.font = Font(name="Arial", size=10)
        for col in ws.columns:
            col_letter = col[0].column_letter
            max_len = max((len(str(c.value)) if c.value else 0) for c in col)
            ws.column_dimensions[col_letter].width = max_len + 2
    wb.save(output_file)
    print(f"✅ Excel généré : {output_file}")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

def idf_material_settings_to_excel(input_string: str, output_file: str, group_size: int = 5):
    """
    Parse 'MaterialProperty:HeatAndMoistureTransfer:Settings' depuis une chaîne IDF
    et exporte un Excel avec un onglet par groupe de `group_size` objets.
    Champs: Name, Porosity, Initial Water Content Ratio.
    Mise en forme: Arial + largeur colonnes ajustée.
    Gère les lignes préfixées par 'IDF,'.
    """
    KEY = "MaterialProperty:HeatAndMoistureTransfer:Settings"

    def _to_float(val):
        s = str(val).strip()
        if s.startswith("."):  # .72 -> 0.72
            s = "0" + s
        try:
            return float(s)
        except ValueError:
            return s  # au cas où

    # Découpage des objets
    rows = [r.strip() for r in input_string.strip().split(";") if r.strip()]
    materials = []

    for row in rows:
        parts = [p.strip() for p in row.split(",")]

        # Recherche de l'index du KEY (gère 'IDF,KEY,...' et 'KEY,...')
        try:
            key_idx = parts.index(KEY)
        except ValueError:
            # Cas particulier: ligne commençant par 'IDF' puis KEY
            if parts and parts[0].upper() == "IDF" and len(parts) > 1 and parts[1] == KEY:
                key_idx = 1
            else:
                # Ligne non conforme: on ignore proprement
                continue

        # Contrôle de longueur minimale
        if len(parts) < key_idx + 4:
            continue

        name   = parts[key_idx + 1]
        poro   = _to_float(parts[key_idx + 2])
        iwcr   = _to_float(parts[key_idx + 3])

        materials.append({
            "Name": name,
            "Porosity": poro,
            "Initial Water Content Ratio": iwcr
        })

    if not materials:
        # Fichier vide mais valide si aucune donnée
        with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
            pd.DataFrame({"Field": [], "Obj1": []}).to_excel(writer, sheet_name="Obj1-5", index=False)
        return

    # Écriture par blocs dans des onglets distincts
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for start in range(0, len(materials), group_size):
            block = materials[start:start + group_size]

            data = {"Name": [], "Porosity": [], "Initial Water Content Ratio": []}
            for m in block:
                data["Name"].append(m["Name"])
                data["Porosity"].append(m["Porosity"])
                data["Initial Water Content Ratio"].append(m["Initial Water Content Ratio"])

            obj_labels = [f"Obj{j+1}" for j in range(start, start + len(block))]
            df = pd.DataFrame(data, index=obj_labels).T
            df.index.name = "Field"

            sheet_name = f"Obj{start+1}-{start+len(block)}"
            df.to_excel(writer, sheet_name=sheet_name)

    # Mise en forme (Arial + auto-width) sur chaque onglet
    wb = load_workbook(output_file)
    for ws in wb.worksheets:
        for row in ws.iter_rows():
            for cell in row:
                cell.font = Font(name="Arial", size=10)
        for col in ws.columns:
            col_letter = col[0].column_letter
            max_len = max((len(str(c.value)) if c.value is not None else 0) for c in col)
            ws.column_dimensions[col_letter].width = max_len + 2
    wb.save(output_file)
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

def idf_sorption_to_excel(input_string: str, output_file: str, group_size: int = 5):
    """
    Parse les objets 'MaterialProperty:HeatAndMoistureTransfer:SorptionIsotherm'
    et exporte un Excel structuré.
    
    Colonnes : HR_1, MC_1, HR_2, MC_2, ...
    Gère 'IDF,' en préfixe.
    Regroupe par groupes de `group_size` dans des onglets séparés.
    """
    KEY = "MaterialProperty:HeatAndMoistureTransfer:SorptionIsotherm"

    # Split et parse
    rows = [r.strip() for r in input_string.strip().split(";") if r.strip()]
    materials = []

    for row in rows:
        parts = [p.strip() for p in row.split(",")]

        # Cas où la ligne commence par "IDF,"
        if parts[0].upper() == "IDF":
            parts = parts[1:]

        if not parts or parts[0] != KEY:
            continue

        name = parts[1]
        try:
            n_points = int(parts[2])
        except ValueError:
            n_points = 0

        values = parts[3:]
        # Paires HR, MC
        hr_mc = []
        for i in range(0, len(values), 2):
            if values[i] == "":
                continue
            try:
                hr = float(values[i])
                mc = float(values[i+1]) if i+1 < len(values) else None
                hr_mc.append((hr, mc))
            except ValueError:
                continue

        materials.append({
            "Name": name,
            "nPoints": n_points,
            "Pairs": hr_mc
        })

    # Transformer en DataFrame pour Excel
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for start in range(0, len(materials), group_size):
            block = materials[start:start + group_size]

            rows_out = {}
            rows_out["Name"] = [m["Name"] for m in block]
            rows_out["nPoints"] = [m["nPoints"] for m in block]

            # max nombre de couples
            max_pairs = max(len(m["Pairs"]) for m in block)

            for j in range(max_pairs):
                rows_out[f"HR_{j+1}"] = [m["Pairs"][j][0] if j < len(m["Pairs"]) else None for m in block]
                rows_out[f"MC_{j+1}"] = [m["Pairs"][j][1] if j < len(m["Pairs"]) else None for m in block]

            df = pd.DataFrame(rows_out, index=[f"Obj{i+1}" for i in range(start, start+len(block))]).T
            df.index.name = "Field"

            sheet_name = f"Obj{start+1}-{start+len(block)}"
            df.to_excel(writer, sheet_name=sheet_name)

    # Mise en forme
    wb = load_workbook(output_file)
    for ws in wb.worksheets:
        for row in ws.iter_rows():
            for cell in row:
                cell.font = Font(name="Arial", size=10)
        for col in ws.columns:
            col_letter = col[0].column_letter
            max_len = max((len(str(c.value)) if c.value else 0) for c in col)
            ws.column_dimensions[col_letter].width = max_len + 2
    wb.save(output_file)
    print(f"✅ Excel généré : {output_file}")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Font

def idf_suction_to_excel(input_string: str, output_file: str, group_size: int = 5):
    KEY = "MaterialProperty:HeatAndMoistureTransfer:Suction"

    rows = [r.strip() for r in input_string.strip().split(";") if r.strip()]
    materials = []

    for row in rows:
        parts = [p.strip() for p in row.split(",")]
        if parts[0].upper() == "IDF":
            parts = parts[1:]

        if not parts or parts[0] != KEY:
            continue

        name = parts[1]
        try:
            n_points = int(parts[2])
        except ValueError:
            n_points = 0

        values = parts[3:]
        wc_suction = []
        for i in range(0, len(values), 2):
            if values[i] == "":
                continue
            try:
                wc = float(values[i])
                suction = float(values[i+1]) if i+1 < len(values) else None
                wc_suction.append((wc, suction))
            except ValueError:
                continue

        materials.append({"Name": name, "nPoints": n_points, "Pairs": wc_suction})

    # --- sécurité : si aucun matériau n'est trouvé ---
    if not materials:
        df = pd.DataFrame({"Info": ["Aucune donnée trouvée"]})
        df.to_excel(output_file, sheet_name="Empty", index=False)
        return

    # Export Excel
    with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
        for start in range(0, len(materials), group_size):
            block = materials[start:start + group_size]

            rows_out = {"Name": [m["Name"] for m in block],
                        "nPoints": [m["nPoints"] for m in block]}

            max_pairs = max(len(m["Pairs"]) for m in block)
            for j in range(max_pairs):
                rows_out[f"WC_{j+1}"] = [m["Pairs"][j][0] if j < len(m["Pairs"]) else None for m in block]
                rows_out[f"Suction_{j+1}"] = [m["Pairs"][j][1] if j < len(m["Pairs"]) else None for m in block]

            df = pd.DataFrame(rows_out, index=[f"Obj{i+1}" for i in range(start, start+len(block))]).T
            df.index.name = "Field"

            sheet_name = f"Obj{start+1}-{start+len(block)}"
            df.to_excel(writer, sheet_name=sheet_name)

    # Mise en forme
    wb = load_workbook(output_file)
    for ws in wb.worksheets:
        for row in ws.iter_rows():
            for cell in row:
                cell.font = Font(name="Arial", size=10)
        for col in ws.columns:
            col_letter = col[0].column_letter
            max_len = max((len(str(c.value)) if c.value else 0) for c in col)
            ws.column_dimensions[col_letter].width = max_len + 2
    wb.save(output_file)
    print(f"✅ Fichier Excel généré : {output_file}")
```
